```python
SELECT * FROM 'flights.csv';
```

```python
-- Average delay for each airline
SELECT name AS airline, ROUND(AVG(arr_delay+dep_delay),0) AS avg_delay
FROM df
WHERE arr_delay IS NOT NULL AND dep_delay IS NOT NULL
GROUP BY airline
ORDER BY avg_delay DESC
```

```python
--What percentage of flights for a given origin destination have delays of more than 15 minutes?
SELECT origin,
	COUNT(*) AS total_flights,
	SUM(CASE WHEN arr_delay + dep_delay > 15 THEN 1 ELSE 0 END) AS delayed_flights,
	ROUND((100.0 * delayed_flights) / COUNT(*),2) AS delay_percent
FROM df
WHERE arr_delay IS NOT NULL AND dep_delay IS NOT NULL
GROUP BY origin
ORDER BY delay_percent DESC
```

```python
--What are the most frequently used routes?
SELECT origin, dest, COUNT(*) AS flights
FROM df
GROUP BY origin, dest
ORDER BY flights DESC
```

```python
-- Which routes have the longest average flight time?
SELECT origin, dest, ROUND(AVG(air_time),0) AS avg_airtime
FROM df
GROUP BY origin,dest
ORDER BY avg_airtime DESC
```

```python
--Which carrier operates the most flights?
SELECT carrier, COUNT(*) AS flights
FROM df
GROUP BY carrier
ORDER BY flights DESC
```

```python
--Are there any airlines that are more likely to land earlier than planned?
SELECT name AS airline, AVG(arr_delay) AS avg_delay
FROM df
WHERE arr_delay<0
GROUP BY airline
ORDER BY avg_delay
```
