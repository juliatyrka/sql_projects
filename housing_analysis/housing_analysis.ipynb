```python
SELECT * FROM 'flats_data.csv';
```

```python
-- AVG price in each city
SELECT city, ROUND(AVG(price_pln),0) AS avg_price
FROM df
GROUP BY city
ORDER BY avg_price DESC
```

```python
-- AVG price per m² in each city
SELECT city, ROUND(AVG(price_pln / area_m2),0) AS price_per_meter
FROM df
GROUP BY city
ORDER BY price_per_meter DESC
```

```python
-- AVG price per m² in each district
SELECT city, district, ROUND(AVG(price_pln / area_m2),0) AS price_per_meter
FROM df
GROUP BY city,district
ORDER BY price_per_meter DESC
```

```python
-- What is the difference in the average price per m² between the primary and secondary market?
WITH avg_prices AS (
  SELECT 
    market_type,
    AVG(price_pln / area_m2) AS avg_price_m2
  FROM df
  GROUP BY market_type
)

SELECT 
  MAX(CASE WHEN market_type = 'primary' THEN avg_price_m2 END) -
  MAX(CASE WHEN market_type = 'secondary' THEN avg_price_m2 END) AS difference_primary_vs_secondary
FROM avg_prices;
```

```python
--How many apartments on the secondary market have more than 3 rooms and cost less than PLN 700,000?
SELECT city, district, COUNT(*) AS result
FROM df
WHERE market_type = 'secondary' AND rooms > 3 AND price_pln < 700000
GROUP BY city, district
```

```python
--Which district in Warsaw has the highest number of ads?
SELECT city, district, COUNT(*) AS result
FROM df
WHERE city = 'Warszawa'
GROUP BY city,district
```

```python
--What is the average price per m² in Warsaw districts?
SELECT city, district, ROUND(AVG(price_pln / area_m2),0) AS price_per_meter
FROM df
WHERE city = 'Warszawa'
GROUP BY city, district
ORDER BY price_per_meter DESC
```

```python
--How many apartments in Warsaw cost less than PLN 600,000 and have at least 2 rooms?
SELECT district,COUNT(*) AS result
FROM df
WHERE price_pln < 600000 AND rooms>=2 AND city='Warszawa'
GROUP BY district
ORDER BY result DESC
```
