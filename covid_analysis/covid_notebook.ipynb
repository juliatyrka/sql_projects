```python
SELECT * FROM 'country_wise_latest.csv';
```

```python
--Which countries have the highest number of confirmed cases?
SELECT region, confirmed
FROM covid
ORDER BY confirmed DESC
LIMIT 10
```

```python
--Which countries have the highest number of deaths?
SELECT region, deaths
FROM covid
ORDER BY deaths DESC
LIMIT 10
```

```python
--Which countries have the highest number of recovered cases?
SELECT region, recovered
FROM covid
ORDER BY recovered DESC
LIMIT 10
```

```python
--What is the death rate for each country?
SELECT region, deaths, confirmed, 
	ROUND((deaths / confirmed),2) AS death_rate
FROM covid
ORDER BY death_rate DESC
```

```python
--Which countries have the highest number of active cases?
SELECT region, active
FROM covid
ORDER BY active DESC
LIMIT 10
```

```python
--What is the total number of cases (confirmed, deaths, recovered, active)?
SELECT region, confirmed 
FROM covid
ORDER BY confirmed DESC
```

```python
--What is the total number of cases worldwide?
SELECT SUM(confirmed) AS result
FROM covid
```

```python
--Which countries have the highest recovery rate?
SELECT region, ROUND(recovered / confirmed,3) AS recovery_rate
FROM covid
WHERE recovery_rate>0.9
ORDER BY recovery_rate DESC
```

```python
--Which countries have more deaths than recoveries?
SELECT region, deaths, recovered
FROM covid
WHERE deaths > recovered
ORDER BY (deaths-recovered) DESC
```

```python
--What percentage of global confirmed cases is in each country?
SELECT region,
       confirmed,
       ROUND(confirmed * 100.0 / (SELECT SUM(confirmed) FROM covid), 2) AS global_share
FROM covid
ORDER BY global_share DESC
```
