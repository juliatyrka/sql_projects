```python
SELECT * FROM "sport_car_prices_cleaned.csv"
```

```python
--Which car brand has the most models in the dataset?
SELECT Brand, COUNT(*) AS cars
FROM cars
GROUP BY Brand
ORDER BY cars DESC
LIMIT 5
```

```python
--What is the average price of cars by brand?
SELECT Brand, 
	ROUND(AVG(Price), 0) AS avg_price_in_thousands
FROM cars
GROUP BY Brand
ORDER BY avg_price_in_thousands DESC
```

```python
--What is the average horsepower and torque by brand?
SELECT Brand, ROUND(AVG(horsepower),2) AS avg_horsepower, ROUND(AVG(torque),2) AS avg_torque
FROM cars
GROUP BY Brand
ORDER BY avg_horsepower DESC, avg_torque DESC
```

```python
--Which 5 models have the highest horsepower?
SELECT Brand, Model, HorsePower
FROM cars
ORDER BY horsepower DESC
LIMIT 5
```

```python
--Which cars have the best acceleration (lowest 0-60 MPH time)?
SELECT Brand, Model, "0_60_MPH_Time"
FROM cars
GROUP BY Brand, Model, "0_60_MPH_Time"
ORDER BY "0_60_MPH_Time"
```

```python
--What are the 3 most expensive models?
SELECT Brand, Model, Price AS Price_in_thousands
FROM cars
GROUP BY Brand, Model, Price_in_thousands
ORDER BY Price_in_thousands DESC
LIMIT 3
```

```python
--Which cars have over 600 HP and cost less than 200k?
SELECT Brand, Model, Horsepower, Price AS Price_in_thousands
FROM cars
WHERE horsepower>600 AND Price_in_thousands<200
ORDER BY price_in_thousands
```

```python
--Rank brands by average 0-60 MPH time
SELECT Brand,
	RANK() OVER(ORDER BY AVG("0_60_MPH_Time")) AS Rank
FROM cars
GROUP BY Brand
ORDER BY Rank
```

```python
--Which brands offer the best horsepower-to-price ratio?
SELECT Brand,
       Model,
       HorsePower,
       Price,
       ROUND(HorsePower / Price, 2) AS hp_to_price_ratio
FROM cars
GROUP BY Brand, Model, HorsePower, Price
ORDER BY hp_to_price_ratio DESC
```
